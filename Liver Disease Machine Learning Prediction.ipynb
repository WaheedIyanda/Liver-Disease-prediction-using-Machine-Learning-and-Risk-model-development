# Training of Machine Learning model

```python
disease_data
```

```python
# separating Target from feaature
Target = disease_data["Diagnosis"]
```

```python

# performing one hot encodement for genetic risk due to having multiple possibiltiy
geneticrisk_encoded = pd.get_dummies(disease_data['GeneticRisk'], prefix='GeneticRisk', dtype=int)
disease_data_encoded = pd.concat([disease_data.drop(columns=['GeneticRisk']), geneticrisk_encoded], axis=1)
```

```python
disease_data_encoded
```

```python
feature = disease_data_encoded.drop(columns=["Diagnosis"])
```

```python
feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature.head(10)
```

```python
feature.columns
```

```python
X= feature
y= Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#using decision tree model
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier(criterion="entropy", max_depth = 5)
```

```python
dtree.fit(X_train, y_train)
```

```python
prediction= dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print (confusion_matrix(y_test, prediction))
```

```python
print (classification_report(y_test, prediction))
```

```python
def get_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

# Function to collect input for each feature
def collect_input_features():
    print("Please enter the following information:")

    # Initialize an empty dictionary to store input features
    input_features = {}

    # Get input for each feature
    input_features['AGE'] = get_input("Age: ")
    input_features['gender'] = get_input("Gender (1 for male, 0 for female): ")
    input_features['BMI'] = get_input("BMI: ")
    input_features['AlcoholConsumption'] = get_input("Alcohol Consumption (scale 0-5): ")
    input_features['Smoking'] = get_input("Smoking (scale 0-3): ")
    input_features['PhysicalActivity'] = get_input("Physical Activity (scale 0-3): ")
    input_features['Diabetes'] = get_input("Diabetes (0 for no, 1 for yes): ")
    input_features['Hypertension'] = get_input("Hypertension (0 for no, 1 for yes): ")
    input_features['LiverFunctionTest'] = get_input("Liver Function Test (0-1): ")
    input_features['GeneticRisk'] = get_input("Genetic Risk Score (0-1): ")

    return input_features

# Main function to run the prediction process
def main():
    # Collect input features from the user
    input_features = collect_input_features()

    # Convert input features into an array
    input_array = np.array([
        input_features['AGE'],
        input_features['gender'],
        input_features['BMI'],
        input_features['AlcoholConsumption'],
        input_features['Smoking'],
        input_features['PhysicalActivity'],
        input_features['Diabetes'],
        input_features['Hypertension'],
        input_features['LiverFunctionTest'],
        input_features['GeneticRisk']
    ])

    # Reshape the array into a format suitable for prediction (1 row, and however many columns are needed)
    input_reshape = input_array.reshape(1, -1)

    # Make prediction using the loaded model (clf)
    prediction = dtree.predict(inpute_reshape)

    # Print the prediction result
    if prediction == 0:
        print('Low Risk of liver disease')
    elif prediction == 1:
        print('High risk of liver disease')
    else:
        print('Not applicable')

# Run the main function
if __name__ == "__main__":
    main()
```
